#### 파일 하나만 처리

````python
READ_DIR = './DATA/Task01_BrainTumour/labelsTr' # label nii.gz가 있는 디렉토리
WRITE_DIR = './DATA/new_processing_brain_png_data/new_label_all'
NII_FILE = 'BRATS_001.nii.gz' # 읽어올 파일명
WRITE_FILE_NAME = None # 밑에서 설정
````

````python
# 설정한 WRITE_DIR 초기화
# !!!!!!!!!!! 파일 삭제 조심 !!!!!!!!!
# 경로설정 : WRITE_DIR 경로 존재할경우 기존존재하는 경로와 파일 지운 후 경로 다시 생성
if os.path.isdir(WRITE_DIR) :
    print("***** 기존에 경로존재, 기존경로와 파일 삭제 후 새로운 경로생성 *****")
    print(WRITE_DIR)
    shutil.rmtree(WRITE_DIR) # 경로파일 모두 삭제
    os.mkdir(WRITE_DIR) # 경로설정
else :
    print("***** 새로운 경로생성 *****")
    print(WRITE_DIR)
    os.mkdir(WRITE_DIR) # 경로설정
# x = shutil.rmtree(WRITE_DIR) and os.mkdir(WRITE_DIR) if os.path.isdir(WRITE_DIR) else os.mkdir(WRITE_DIR) 
````

````python
# https://nipy.org/nibabel/reference/nibabel.dataobj_images.html#module-nibabel.dataobj_images

img = nib.load(os.path.join(READ_DIR, NII_FILE)) # nii 불러오기

# get_fdata시 default 인 float으로 가져옴
# data = img.get_fdata(dtype=np.uint8) # 불러온 nii numpy로 변환, float32
# data_uint = data[:,:,:].astype(np.uint8) # data numpy 정수화
# print("data uint shape: ", data_uint.shape)
# print("data uint unique", np.unique(data_uint))

# np.asarray(prox) 로 불러오면 dataobj dtype으로 불러옴
# 불러온 img 객체에 dataobj객체가 진짜 데이터 이미지 정보, 불러올때 저장한 타입으로 불러옴 즉,  uint8
# data_obj = img.get_data() # 이건 되는지 잘 몰라
# print(img.dataobj) # 이건 되는지 잘 몰라

# img_numpy = 불러온 nii 파일의 numpy data
img_obj = img.dataobj # arrayproxy object
img_numpy = np.asarray(img_obj) # arrayproxy numpy로 변환

print("img_numpy ==> dtype:{} | shape:{} | nuique:{} ← img_numpy 전체 원소(label)정보" .format(img_numpy.dtype, img_numpy.shape, np.unique(img_numpy)))
print("----------------\n")

# z축과 label값 설정
z = 77 # 잘라올 z축
label_value = 3 # enhancing tumour - t1GD
print("z = {}, label_value = {}".format(z, label_value))
print("np.unique(img_numpy[:,:,z]) ==> unique: {} ← img_numpy[:,:,z] 원소(label)정보".format(np.unique(img_numpy[:,:,z])))
print("----------------\n")

# img_z_label_boolean = 불러온 img에 z축 적용, label_value적용, boolean 타입
img_z_label_boolean = (img_numpy[:,:,z] == label_value).copy() # nii 파일의 z축 잘라오기, 찾아올 label값도 설정
print("img_z_label_boolean ==> dtype:{} | shape:{}".format(img_z_label_boolean.dtype, img_z_label_boolean.shape))
print("----------------\n")

# img_z_label_boolean 출력
plt.imshow(img_z_label_boolean)

# 3채널 8비트 만들기 == 저장할 이미지
img_save = img_z_label_boolean.astype(np.uint8) * 255 # 저장할 img
print("img_save ==> dtype:{} | shape:{} | unique:{} ← 8비트로 변경".format(img_save.dtype, img_save.shape, np.unique(img_save)))

img_save = cv2.cvtColor(img_save, cv2.COLOR_GRAY2RGB) # rgb 채널로 변경
img_save = np.rot90(img_save) # 반시계 90도 회전
print("img_save ==> dtype:{} | shape:{} | unique:{} ← GRAY -> RGB (3채널로 변경)".format(img_save.dtype, img_save.shape, np.unique(img_save)))
print("----------------\n")

# 저장
# WRITE_FILE_NAME = "{}-label{}-slice{}.jpg".format(NII_FILE.split('.')[0], label_value, z) # 저장파일 이름설정

# print("SAVING ------------------------> " + WRITE_DIR +'/'+ WRITE_FILE_NAME)
# cv2.imwrite(os.path.join(WRITE_DIR, WRITE_FILE_NAME), img_save) # write하기 위해 이미 path가 설정되어 있어야 함
````

#### 디렉토리에 있는 모든 파일 한번에 처리

````python
# https://nipy.org/nibabel/reference/nibabel.dataobj_images.html#module-nibabel.dataobj_images

READ_DIR = './DATA/Task01_BrainTumour/labelsTr' # label nii.gz가 있는 디렉토리
WRITE_DIR = './DATA/new_processing_brain_data/new_label' # label 처리후 저장할 디렉토리
WRITE_FILE_NAME = None # 밑에서 설정
fcount = 0 # 처리한 파일 개수

file_list = [file for file in os.listdir(READ_DIR) if file.startswith("BRATS")] # BRATS로 시작되는 파일명만 가져오기

print("{} FILES DETECTED \n".format(len(file_list)))

for fname in file_list : # read_dir에 있는 모든 파일 가져오기
    
    print("Loop.{}".format(fcount))
    print("TARGET FILE: {}".format(fname)) # target file
    
    img = nib.load(os.path.join(READ_DIR, fname)) # nii 불러오기

    img_obj = img.dataobj # arrayproxy object
    img_numpy = np.asarray(img_obj) # arrayproxy numpy로 변환

#    print("img_numpy ==> dtype:{} | shape:{} | nuique:{} ← img_numpy 전체 원소(label)정보" .format(img_numpy.dtype, img_numpy.shape, np.unique(img_numpy)))
#    print("----------------\n")

    # z축과 label값 설정
    z = 77 # 잘라올 z축
    label_value = 3 # enhancing tumour - t1GD
#    print("z = {}, label_value = {}".format(z, label_value))
#    print("np.unique(img_numpy[:,:,z]) ==> unique: {} ← img_numpy[:,:,z] 원소(label)정보".format(np.unique(img_numpy[:,:,z])))
#    print("----------------\n")

    # img_z_label_boolean = 불러온 img에 z축 적용, label_value적용, boolean 타입
    img_z_label_boolean = (img_numpy[:,:,z] == label_value).copy() # nii 파일의 z축 잘라오기, 찾아올 label값도 설정
#    print("img_z_label_boolean ==> dtype:{} | shape:{}".format(img_z_label_boolean.dtype, img_z_label_boolean.shape))
#    print("----------------\n")

    # img_z_label_boolean 출력
#    plt.imshow(img_z_label_boolean)

    # 3채널 8비트 만들기 == 저장할 이미지
    img_save = img_z_label_boolean.astype(np.uint8) * 255 # 저장할 img
    print("img_save ==> dtype:{} | shape:{} | unique:{} ← 8비트로 변경".format(img_save.dtype, img_save.shape, np.unique(img_save)))
    
#    img_save = cv2.cvtColor(img_save, cv2.COLOR_GRAY2RGB) # rgb 채널로 변경
#     print("img_save ==> dtype:{} | shape:{} | unique:{} ← GRAY -> RGB (3채널로 변경)".format(img_save.dtype, img_save.shape, np.unique(img_save)))
#     print("----------------\n")
    
    img_save = np.rot90(img_save) # 반시계 90도 회전
#     img_save = img_save[..., np.newaxis] # 차원확장
    print("img_save ==> dtype:{} | shape:{} | unique:{} ← 반시계 90도 회전, 저장 전 최종데이터".format(img_save.dtype, img_save.shape, np.unique(img_save)))
    
#    plt.imshow(img_save)
       
    # 저장 : cv2.imwrite의 경우 3채널로 확장해 저장하는 듯, 저장 후 다시 읽어보니 3채널이였음, 근데 label의 경우 모든 채널 값 분포 같았음
    WRITE_FILE_NAME = "{}-label{}-slice{}.png".format(fname.split('.')[0], label_value, z) # 저장파일 이름설정
    
    cv2.imwrite(os.path.join(WRITE_DIR, WRITE_FILE_NAME), img_save) # write하기 위해 이미 path가 설정되어 있어야 함
    
    fcount += 1
    
    print("SAVING ------------------------> " + WRITE_DIR +'/'+ WRITE_FILE_NAME)
    
print("\n@ {} FILES SAVED".format(fcount))
````

````python
# https://nipy.org/nibabel/reference/nibabel.dataobj_images.html#module-nibabel.dataobj_images

img = nib.load(os.path.join(READ_DIR, NII_FILE)) # nii 불러오기

# get_fdata시 default 인 float으로 가져옴
# data = img.get_fdata(dtype=np.uint8) # 불러온 nii numpy로 변환, float32
# data_uint = data[:,:,:].astype(np.uint8) # data numpy 정수화
# print("data uint shape: ", data_uint.shape)
# print("data uint unique", np.unique(data_uint))

# np.asarray(prox) 로 불러오면 dataobj dtype으로 불러옴
# 불러온 img 객체에 dataobj객체가 진짜 데이터 이미지 정보, 불러올때 저장한 타입으로 불러옴 즉,  uint8
# data_obj = img.get_data() # 이건 되는지 잘 몰라
# print(img.dataobj) # 이건 되는지 잘 몰라

# img_numpy = 불러온 nii 파일의 numpy data
img_obj = img.dataobj # arrayproxy object
img_numpy = np.asarray(img_obj) # arrayproxy numpy로 변환

print("img_numpy ==> dtype:{} | shape:{} | nuique:{} ← img_numpy 전체 원소(label)정보" .format(img_numpy.dtype, img_numpy.shape, np.unique(img_numpy)))
print("----------------\n")


for i in range(img_obj.shape[2]) : 
    
    # z축과 label값 설정
    z = i # 잘라올 z축
    label_value = 3 # enhancing tumour - t1GD
    print("z = {}, label_value = {}".format(z, label_value))
    print("np.unique(img_numpy[:,:,z]) ==> unique: {} ← img_numpy[:,:,z] 원소(label)정보".format(np.unique(img_numpy[:,:,z])))
    print("----------------\n")

    # img_z_label_boolean = 불러온 img에 z축 적용, label_value적용, boolean 타입
    img_z_label_boolean = (img_numpy[:,:,z] == label_value).copy() # nii 파일의 z축 잘라오기, 찾아올 label값도 설정
    print("img_z_label_boolean ==> dtype:{} | shape:{}".format(img_z_label_boolean.dtype, img_z_label_boolean.shape))
    print("----------------\n")

    # img_z_label_boolean 출력
    plt.imshow(img_z_label_boolean)

    # 3채널 8비트 만들기 == 저장할 이미지
    img_save = img_z_label_boolean.astype(np.uint8) * 255 # 저장할 img
    print("img_save ==> dtype:{} | shape:{} | unique:{} ← 8비트로 변경".format(img_save.dtype, img_save.shape, np.unique(img_save)))

    img_save = cv2.cvtColor(img_save, cv2.COLOR_GRAY2RGB) # rgb 채널로 변경
    img_save = np.rot90(img_save) # 반시계 90도 회전
    print("img_save ==> dtype:{} | shape:{} | unique:{} ← GRAY -> RGB (3채널로 변경)".format(img_save.dtype, img_save.shape, np.unique(img_save)))
    print("----------------\n")

    # 저장

    WRITE_FILE_NAME = "{}-label{}-slice{}.jpg".format(NII_FILE.split('.')[0], label_value, z) # 저장파일 이름설정
    print("SAVING ------------------------> " + WRITE_DIR +'/'+ WRITE_FILE_NAME)
    cv2.imwrite(os.path.join(WRITE_DIR, WRITE_FILE_NAME), img_save) # write하기 위해 이미 path가 설정되어 있어야 함
````
